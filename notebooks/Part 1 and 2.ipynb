We will start the basic exploration with first loading the set and having a look at it.

```python
import pandas as pd

try:
    df = pd.read_csv('../data/breast_cancer.csv')
except FileNotFoundError as e:
    print(f"Error: {e}")
    exit()
print(df.shape)
print(df.dtypes)
print(df.head())
df.describe()
```

Then we remove duplicates and entries with null values.

```python
print("Duplicate Values:\n")
print(df.duplicated().sum())
print("Missing Values:\n")
print(df.isnull().sum())
# Remove any rows with missing values
df.dropna(inplace=True) # Will not do something

# Remove any duplicate rows
df.drop_duplicates(inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)

df.describe()
```

We see that combaning all the rows with at least one value null quite some data were deleted:
From 512 ids we are down to 384. In another senario we could also fill the gaps with mean values or another method.


Now we want to convert the categorical column "diagnosis" with one-hot encoding to "cancer positive diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

diagnosisEnc = LabelEncoder()

print("Before:\n")
display(df.head())

df["diagnosis"] = diagnosisEnc.fit_transform(df["diagnosis"])

print("After\n")
display(df.head())
```

We want to see if the two classes (malignant - cancerous -1/benign - non-cancerous - 0) are balanced.

```python
print("Class imbalance:\n")
print(df['diagnosis'].value_counts(normalize=True))
```

To view the outliers we want to see the feats that score high and low in IQR.

```python
import numpy as np
def ciqr(dff, col):
    Q1 = dff[col].quantile(0.25)  # 25th percentile
    Q3 = dff[col].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1
    return IQR
    
iqr = {col: ciqr(df, col) for col in df.select_dtypes(include=np.number).columns}
iqr
```

We see which arε the outliers:

```python
{key: value for key, value in iqr.items() if value > 3 and key not in ["diagnosis", "id"]}
```

The ones really standing out are "area_mean" and "area_worst"

We proceed to further feature assessment and visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns

features = df.columns.drop('diagnosis').drop('id')
plt.figure(figsize=(20, 10))
df[features].boxplot(rot=90)
plt.title('Boxplots of Feature Distribution')
plt.show()

plt.figure(figsize=(16, 12))
sns.heatmap(df[features].corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()
```

We see a lot oh highly correlated components (f.e.: area_worst -> perimeter_worst, radius_mean -> perimeter_mean) which is to expected due to the nature of our data. 

We will also try to reduce the dimension of our df seeing a scatter plot of the first two principal components.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#featVal = df[features].values
# To do the pca we first need to stantarize - scale our data
scaled = StandardScaler().fit_transform(df[features])
pca = PCA(n_components=2)
pcaX = pca.fit_transform(scaled)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pcaX[:, 0], y=pcaX[:, 1], hue=df['diagnosis'])
plt.title('PCA: First Two Principal Components')
#plt.xlabel(pcaX)
plt.show()

import numpy as np

# Create a DataFrame of loadings
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(pca.n_components_)],
    index=features
)

# For each component, get the features with the largest absolute values
for pc in loadings.columns:
    print(f"\nTop features for {pc}:")
    print(loadings[pc].abs().sort_values(ascending=False).head(5))

featImportance = loadings.abs().sum(axis=1).sort_values(ascending=False)
print("\nOverall most important features:")
print(featImportance.head(10))

pca2 = PCA(n_components=30)
pcaX2 = pca2.fit_transform(scaled)

cummVar = np.cumsum(pca2.explained_variance_ratio_)
comp95 = np.argmax(cummVar >= 0.95) + 1
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cummVar)+1), cummVar, marker='o', linestyle='--', color='green')
plt.axhline(y=0.95, color='red', linestyle=':', label='95% Explained Variance')
plt.axvline(x=comp95, color='blue', linestyle=':', label=f'{comp95} Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by PCA Components')
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
print(f"Number of components to explain at least 95% variance: {comp95}")
```

From the plot above we see that with PCA we already get a really good idea of what is happening. We have two clusters with only some data being on the edge of them.

As we also saw we had some really correlated features we will try to create a custom reduced df by dropping everything that has a correlation over a threshold (saw the bonus part after I did this already). 
From the Cumulative Variance table we see that we already hit a 95% percent variance from the first 10 components.

```python
# X_imputed: your imputed feature matrix,
corrMat = pd.DataFrame(df, columns=features).corr().abs()
upper = corrMat.where(np.triu(np.ones(corrMat.shape), k=1).astype(bool))
highCorrs = [column for column in upper.columns if any(upper[column] > 0.95)]
reducedFeats = [feature for feature in features if feature not in highCorrs]

print("Highly correlated features to drop:", highCorrs)
print("Reduced feature set:", reducedFeats)

reddf = df[reducedFeats]
display(reddf.head())
# reddf.info()
reddf.describe()

```
